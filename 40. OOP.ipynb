```python
def new_game():
    
    tebakan = []
    tebakan_benar = 0
    pertanyaan_num = 1

    for key in pertanyaan:
        print('-------------------------')
        print(key)
        for i in pilihan[pertanyaan_num-1]:
            print(i)
        
        opsi=['A', 'B', 'C', 'D']
        tebak=None
        while tebak not in opsi:
            tebak = input('Enter (A, B, C, D): ')
            tebak = tebak.upper()
            
        tebakan.append(tebak)

        tebakan_benar += check_answer(pertanyaan.get(key), tebak)
        pertanyaan_num += 1

    display_score(tebakan_benar, tebakan)


# -------------------------
def check_answer(answer, tebak):

    if answer == tebak:
        print('BENAR!')
        return 1
    else:
        print('SALAH!')
        return 0
    
# -------------------------
def display_score(tebakan_benar, tebakan):
    print('-------------------------')
    print('HASIL')
    print('-------------------------')
    
    print('Jawaban: ', end='')
    for i in pertanyaan:
        print(pertanyaan.get(i), end=' ')
    print()

    print('Tebakan: ', end='')
    for i in tebakan:
        print(i, end=' ')
    print()

    score = int((tebakan_benar/len(pertanyaan))*100)
    print(f"Score anda adalah: {str(score)}%")


# -------------------------
def play_again():
    
    response = input('Mau bermain lagi? (Y/N)').lower()

    if response == 'y':
        return True
    else:
        return False

# -------------------------

pertanyaan = {"Ada berapa element di periodic table?: ": 'A',
              "Binatang apa yang memiliki telur terbesar?: ": 'B',
              "Binatang yang memiliki 4 kaki?: ": 'C',
              "Ada berapa banyak tulang manusia dewasa?: ": 'A',
              "Planet mana yang terpanas di tata surya?: ": 'C'
              } # pasangan pertanyaan dan jawaban yang benar


pilihan = [['A. 118', 'B. 120', 'C. 127', 'D. 133'], 
           ['A. ayam', 'B. burung unta', 'C. burung elang', 'D. biawak'], 
           ['A. laba-laba', 'B. ayam', 'C. kuda', 'D. hiu'], 
           ['A. 206', 'B. 270', 'C. 208', 'D. 214'], 
           ['A. bumi', 'B. uranus', 'C. merkurius', 'D. venus']]




new_game()


while play_again():
    new_game()

print('BYEEEE!')
```
